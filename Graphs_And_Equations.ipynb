```python
import seaborn as sb
import numpy as np
import pandas as pd
```

```python
####################
# LINEAR EQUATIONS #
####################

#############
# m = slope #
#############

###########################
# m = (y2 - y) / (x2 - y) #
###########################


def generate_points(x_i, y_i, m, n_points) -> pd.DataFrame:
    dict = { "x" : [], "y" : [] }
    curr_x = x_i
    curr_y = y_i
    i = n_points
    while i > 0:
        dict["x"].append(curr_x)
        dict["y"].append(curr_y)
        curr_x += m["x"]
        curr_y += m["y"]
        i -= 1
    data = pd.DataFrame(dict)
    sb.lineplot(data=data, x="x", y="y")
    sb.scatterplot(data=data, x="x", y="y")
    return data
```

```python
##############################
#      f = 4x - 3y + 5       #
##############################
#                            #
# x intercept = (-5/4 ,  0)  #
#                            #
# y intercept = (0 , -5/-3)  #    
#                            #
# m = -4/-3                  #
##############################
data = generate_points((-5/4), (-5/-3), { "x" : -4, "y" : -3 }, 10)
data
```

```python
#######################
# QUADRATIC EQUATIONS #
#######################

#########################
# x intercept = -b / 2a #
#########################

#########################################################
# quadratic formula = ( -b +/- sqrt( b^2 - 4ac ) ) / 2a #
#########################################################

def calc_eq(eq, val) -> int:
    res = (eq["a"] * (val * val)) + (eq["b"] * val) + eq["c"]
    return res

def generate_quadratic_points(eq, num_p) -> pd.DataFrame:
    dict = {"x":[], "y":[]}
    x_i = ( -1 * eq["b"] ) / (2 * eq["a"])

    curr_x = x_i - num_p
    i = num_p * 2 + 1
    while i > 0:
        curr_y = calc_eq(eq, curr_x)
        if i == num_p + 1:
            print("\n\nVertex:", "(", x_i, ",", curr_y, ")\n\n")
        dict["x"].append(curr_x)
        dict["y"].append(curr_y)
        curr_x += 1
        i -= 1
    data = pd.DataFrame(dict)
    sb.lineplot(data=data, x="x", y="y")
    sb.scatterplot(data=data, x="x", y="y")
    return data

    



```

```python
########################
# y = -2x^2 + 12x - 10 #
########################

data = generate_quadratic_points({ "a" : -2, "b" : 12, "c":-10 }, 10)
data
```
